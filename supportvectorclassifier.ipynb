```python
import pandas as pd 
import numpy as np 
from sklearn.svm import SVC 
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import classification_report 
from sklearn.metrics import accuracy_score, confusion_matrix 
from sklearn.model_selection import train_test_split, cross_val_score,StratifiedKFold
```

```python
dataframe = pd.read_csv('diabetes(2).csv')
dataframe
```

```python
array = dataframe.values
x = array[:,0:8]
y = array[:,8]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, stratify = y)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
clf = SVC()
clf.fit(x_train,y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
print(classification_report(y_test,y_predict))
```

```python
accuracy_score(y_train, clf.predict(x_train))
```

### Hyper Parameter Tuning with Randomized Grid Search CV

```python
clf = SVC()
param_grid = [{'kernel':['linear','rbf'],'gamma':[0.1,0.5,1],'C':[0.1,1,10] }]
kfold = StratifiedKFold(n_splits=5)
gsv = RandomizedSearchCV(clf,param_grid,cv=kfold,scoring= 'recall')
gsv.fit(x_train,y_train)
```
